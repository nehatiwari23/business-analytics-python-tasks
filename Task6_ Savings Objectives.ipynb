 Task 6: Savings Objectives

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('dataset.csv')  # replace with the actual file path

# Analyze the 'Savings Objectives' column
savings_objectives_counts = df['What are your savings objectives?'].value_counts()

print(savings_objectives_counts)
```

 Visualizing the distribution of savings objectives using a bar chart:

```python
savings_objectives_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Savings Objectives')
plt.xlabel('Savings Objective')
plt.ylabel('Number of Participants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
