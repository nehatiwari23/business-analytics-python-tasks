Task 7: Common Information Sources

```python
 #Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv("dataset.csv")

# Split and clean
df['Information Sources'] = df['Source'].str.lower()
df['Information Sources'] = df['Source'].str.split(',')


# Flatten list of sources
all_sources = df['Source'].explode().str.strip()
df_exploded = df.explode('Information Sources')

# Count frequency
source_counts = all_sources.value_counts()
print(source_counts)

```

```python


# ✅ Count each individual source
source_counts = df_exploded['Information Sources'].value_counts().reset_index()
source_counts.columns = ['Information Source', 'Count']


# ✅ Bar Chart
plt.figure(figsize=(8, 5))
plt.barh(source_counts['Information Source'], source_counts['Count'], color='mediumseagreen')
plt.xlabel('Number of Participants')
plt.title('Common Sources of Investment Information')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# ✅ Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(
    source_counts['Count'],
    labels=source_counts['Information Source'],
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Distribution of Information Sources')
plt.axis('equal')
plt.tight_layout()
plt.show()

```
