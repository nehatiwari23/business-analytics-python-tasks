```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('dataset.csv')

# List of reason columns to analyze
reason_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

for col in reason_columns:
    print(f"\n--- Analysis for {col} ---")
    
    # Drop missing values
    reasons = df[col].dropna()
    
    # Count all unique reasons exactly as they appear
    summary = reasons.value_counts().reset_index()
    summary.columns = ['Reason', 'Count']
    
    # Print all reasons with counts
    print(summary)
    
    # Plot all reasons — beware, very long lists can be hard to read
    plt.figure(figsize=(12, max(6, 0.3 * len(summary))))  # dynamic height
    plt.barh(summary['Reason'], summary['Count'], color='steelblue')
    plt.xlabel('Count')
    plt.title(f'All Reasons for Investment - {col}')
    plt.gca().invert_yaxis()  # highest count on top
    plt.tight_layout()
    plt.show()
```
