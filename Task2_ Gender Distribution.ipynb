1. Load the Dataset

```python
import pandas as pd

# Load dataset
df = pd.read_csv("dataset.csv")
```

2. Extract Gender Column

```python
print(df.columns) 
gender_data = df["gender"]
```

3. Check Unique Values in Gender

```python
print(gender_data.value_counts())
```

4. Visualize Gender Distribution

a. Bar Chart

```python
import matplotlib.pyplot as plt

gender_counts = gender_data.value_counts()

plt.figure(figsize=(6,4))
gender_counts.plot(kind='bar', color=['skyblue', 'lightpink', 'lightgreen'])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

b. Pie Chart

```python

import matplotlib.pyplot as plt

gender_counts = df['gender'].value_counts()

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, 
        labels=gender_counts.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['skyblue', 'lightpink', 'lightgreen'])
plt.title("Gender Distribution")
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle
plt.show()
```
